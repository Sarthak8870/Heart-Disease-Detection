```python
df4=df3.copy()
df4.ExerciseAngina.replace({
    'N':0,
    'Y':1
}, inplace=True)

df4.ST_Slope.replace({
    'Down':1,
    'Flat':2,
    'Up':3
},inplace=True)

df4.RestingECG.replace({
    'Normal':1,
    'ST':2,
    'LVH':3
},inplace=True)
df4.head()
```

```python
df5=pd.get_dummies(df4, drop_first=True, dtype='int')
df5.head()
```

```python
x=df5.drop('HeartDisease',axis=1)
y=df5.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25, random_state=15)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(x_scaled)
```

```python
x_pca.shape
```

```python
x_train_pca,x_test_pca,y_train_pca,y_test_pca=train_test_split(x_pca,y,test_size=0.25, random_state=15)
```

```python
model.fit(x_train_pca,y_train_pca)
```

```python
model.score(x_test_pca,y_test_pca)
```
